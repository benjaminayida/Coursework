```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **1. Create a numpy array with 100 elements from 1 to 100 in order : use box plot to draw the data you generate**

```python
arr = np.arange(1,101)
plt.boxplot(arr)
plt.show()
```

#### **2. Create a numpy array with 8000 random numbers: display a histogram of the data using 20 bins**

```python
arr = np.random.randn(8000)
plt.hist(arr, bins=20)
plt.show()
```

#### **3. Reshape it into a 20 x 20 x 20 array, and with the reshaped numpy array, compute the average values for every x,y,and z slice**

```python
arr_3D = arr.reshape(20, 20, 20)

x_avg = np.mean(arr_3D, axis=(1, 2))
y_avg = np.mean(arr_3D, axis=(0, 2))
z_avg = np.mean(arr_3D, axis=(0, 1))

print("Average of slice x:\n\n", x_avg,'\n\n',\
      "Average of slice y:\n\n", y_avg,'\n\n',"Average of slice z:\n\n", z_avg,'\n\n')
```

#### **4. Average all the x slices together and display a single 2D image of the averaged x slice. Do the same for y and z slices**

```python
x_avg = np.mean(arr_3D, axis=0)
y_avg = np.mean(arr_3D, axis=1)
z_avg = np.mean(arr_3D, axis=2)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))
axs[0].imshow(x_avg)
axs[0].set_title('X Average')
axs[1].imshow(y_avg)
axs[1].set_title('Y Average')
axs[2].imshow(z_avg)
axs[2].set_title('Z Average')
plt.show()
```

#### **5. Write a python program to generate 100 random numbers uniformly distributed between 1 and 100 , write the numbers out to a binary file : use a line graph to draw the 100 numbers**

```python
rand_num = np.random.uniform(1, 100, 100)

with open('random_numbers.bin', 'wb') as f:
    f.write(rand_num.tobytes())

plt.plot(rand_num)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('100 Random Numbers')
plt.show()
```

#### **6. Write a python program to read the binary file back, divide the range between 1 and 100 into 7 intervals, and calculate the frequency for each interval: display a histogram of your result**

```python
with open('random_numbers.bin', 'rb') as f:
    data = np.frombuffer(f.read(), dtype=np.float64)

intervals = np.linspace(1, 100, num=8)

freq, _ = np.histogram(data, bins=intervals)

plt.hist(data, bins=intervals, edgecolor='k')
plt.xticks(intervals)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Random Numbers Histogram')
plt.show()

print('Frequency for each interval:')
for i in range(len(freq)):
    print(f'\n[{intervals[i]:.1f}, {intervals[i+1]:.1f}) = {freq[i]}')
```

#### **7. Write a python function that takes a list of random numbers as input, compute the mean and the standard distribution: draw a line graph for your data** 

```python
def func(x):
    # convert the list to a numpy array
    data = np.array(x)
    
    # compute the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)
    
    # plot the data as a line graph
    plt.plot(data)
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title(f'Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}')
    plt.show()

numbers = [3, 6, 8, 12, 9, 10, 5, 2, 7, 11]
func(numbers)
```
